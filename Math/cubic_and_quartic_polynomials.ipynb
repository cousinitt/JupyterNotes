Usually, I would just use a numerical algorithm, but this was just too neat for me to pass up. To solve the cubic and the quartic, we start by standardizing the equations. This is not a lecture on the method, but rather a listing of steps. The interested reader is referred to the voluminous literature on polynomials for more details. First, we divide through by the leading coefficient to make the polynomial [monic ](https://en.wikipedia.org/wiki/Monic_polynomial)(first coefficient is 1). Next, we depress the polynomial. This means that a polynomial of degree $n$ will be missing the term of degree $n âˆ’ 1$. This simplifies the analysis.

## Real Roots of the Cubic

The general monic cubic equation is:

$$
x^3 + a x^2 + b x + c = 0.
$$

The next step would be to depress the cubic, which is a generalization of what you may remember as [completing the square](https://en.wikipedia.org/wiki/Completing_the_square). It involves the [binomial theorem](https://en.wikipedia.org/wiki/Binomial_theorem):

$$
(x + g)^3 = x^3 + 3gx^2 + 3g^2 x + g^3.
$$

We then use this relationship to remove the square term in the monic cubic:

$$
x^3 + (b - a^2/3) x + 2 a^3/27 - ab/3 + c = 0.
$$

For the next step, we will need the discriminant of the polynomial. The discriminant is often written as $\Delta$. The discriminant is the product of the differences of the roots. I won't derive this, and for the following steps we will use [Cardano's method](http://www2.trinity.unimelb.edu.au/~rbroekst/MathX/Cubic%20Formula.pdf).

$$
\Delta = -\frac{q^2}{4} - \frac{p^3}{27}.
$$

If $\Delta = 0$, we have three real roots, two of which are repeated:

$$
\begin{aligned}
r_0 &= -2 \bigl(\frac{q}{2}\bigr)^{\frac{1}{3}} - \frac{a}{3} \\
r_1 &= \bigl(\frac{q}{2}\bigr)^{\frac{1}{3}} - \frac{a}{3}.
\end{aligned}
$$

If $\Delta > 0$, we have three distinct real roots:

$$
\begin{aligned}
\theta &= \frac{1}{3}\cos^{-1}\bigl(\sqrt{\frac{-27 q^2}{4 p^3}}\bigr) \\
r_n &= \sqrt{\frac{-4 p}{3}} \cos\bigl(\theta + \frac{2\pi n}{3} \bigr), n \in 0, 1, 2.
\end{aligned}
$$

If $\Delta < 0$, we have one real root:

$$
r_0 = \bigl(-\frac{q}{2} + \sqrt{-\Delta}\bigr)^{\frac{1}{3}} + \bigl(-\frac{q}{2} - \sqrt{-\Delta}\bigr)^{\frac{1}{3}}.
$$
